```python
import pandas as pd

dataset = pd.read_csv("rideshare_kaggle.csv")
dataset
```

```python
#cleaning data
#getting rid of null values
dataset.isnull().sum()
dropped_dataset = dataset.dropna(subset=["price"])
new_df = dropped_dataset.head(1000)

new_df
```

```python
# split into the independent and dependent variables

X_no_intercept = new_df[['hour', 'day', 'distance', 'month', 'temperature', 'precipIntensity']].values
y = new_df['price'].values
```

```python
new_df[['hour', 'day', 'distance', 'month', 'temperature', 'precipIntensity']].head()
```

```python
import numpy as np 
X = np.column_stack([np.ones(X_no_intercept.shape[0], dtype = 'int'), X_no_intercept])
```

```python
# run this line to view the X matrix
X
```

```python
#perform least squares regression

beta_hat = np.linalg.inv(X.T @ X) @ X.T @ y

# run this line to view the model parameters
beta_hat
```

```python
y_hat = X @ beta_hat

y_hat
```

```python
import matplotlib as mp
import matplotlib.pyplot as plt

#plotting linear model
plt.figure()
plt.scatter(y, y_hat)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.show()
```

```python
import statsmodels.api as sm


Xfull = sm.add_constant(X)
# This adds the column of 1s

Y = pd.DataFrame(new_df["price"])
output = sm.OLS(Y, Xfull).fit()
output.summary()
```

# Alternative Method (Using Dummies to improve R Squared)

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf


X = new_df[['hour', 'day', 'distance', 'month', 'temperature', 'precipIntensity', 'source', 'destination', 'cab_type']]

X =  pd.get_dummies(data=X, drop_first=True)

X
```

```python
Y = new_df['price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.4, random_state=101)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
# print the intercept
print(model.intercept_)
```

```python
#predicted outputs based on input test data

predictions = model.predict(X_test)
predictions
```

```python
coeff_parameter = pd.DataFrame(model.coef_,X.columns,columns=['Coefficient'])
coeff_parameter
```

```python
import seaborn as sns

sns.regplot(y_test,predictions)
```
